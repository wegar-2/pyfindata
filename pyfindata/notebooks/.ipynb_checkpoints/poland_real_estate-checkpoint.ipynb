```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
from pyfindata.extractor.gus_monthly_macro_indicators_extractor import GUSMonthlyMacroIndicatorsExtractor
from pyfindata.extractor.nbp_real_estate_prices_extractor import NBPRealEstatePricesExtractor
```

### 1. Load PL Real Estate Prices Data

```python
re_data = NBPRealEstatePricesExtractor().extract()
```

```python
re_data = re_data.loc[:, ("secondary", "transactional", slice(None))]
```

```python
re_data.columns = re_data.columns.droplevel(level=[0, 1])
```

```python
cities = ["bialystok", "wroclaw", "zielona gora", "szczecin", "gdansk", "warszawa", "rzeszow", "lodz", "poznan"]
```

```python
re_data = re_data.loc[:, cities]
```

```python
re_data.index = pd.to_datetime(re_data.index)
```

```python
re_data = re_data / re_data.loc["2019-12-01", :]
```

### 2. Load PL Inflation Time Series

```python
# load inflation dat
inflation = GUSMonthlyMacroIndicatorsExtractor().extract()
inflation = inflation.loc[~inflation.isna().any(axis=1), :]
```

```python
# get price index anchored at Dec 2019
inflation = inflation[["CPI_MoM", "yearmon"]].set_index("yearmon")
inflation["price_index"] = (inflation["CPI_MoM"] / 100).cumprod()
inflation["price_index"] = inflation["price_index"] / inflation.loc["2019-12-01", "price_index"]
inflation = inflation["2011-01-01":]
```

```python
fig_prlvl = plt.figure(figsize=(15, 8))
ax_prlvl = fig_prlvl.add_subplot(1, 1, 1)
ax_prlvl.plot(inflation["price_index"])
ax_prlvl.grid()
ax_prlvl.set_title("Price lvl in PL, Dec 2019 = 100")
```

### 3. Calculate Real Secondary Market Transactional Prices in Select Cities

```python
inflation = inflation[["price_index"]]
```

```python
re_data = pd.merge(left=re_data, right=inflation, left_index=True, right_index=True, how="left")
```

```python
re_data = re_data["2016-01-01":]
```

```python
real_re = re_data.loc[:, cities].copy(deep=True)
```

```python
for city in cities:
    real_re[city] = re_data[city] / re_data["price_index"]
```

```python
fig_rre = plt.figure(figsize=(12, 8))
ax_rre = fig_rre.add_subplot(1, 1, 1)
for city in cities:
    ax_rre.plot(real_re[city], label=city)
ax_rre.grid()
ax_rre.legend()
ax_rre.set_title("Real estate prices adjusted for inflation")
ax_rre.plot()
```
